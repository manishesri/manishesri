```python
#Create a df which consists of only English Movies
english_movies = imdb_data[imdb_data['original_language'] == 'en']

#Create another df which consists of movies whose length is greater than 100mins
long_movies = imdb_data[imdb_data['runtime'] >=100]['imdb_id']
```

```python
long_movies.shape, english_movies.shape
```

## Inner Join to Merge the two dataframes

```python
merged_df = pd.merge(english_movies,long_movies,how = 'inner', left_on = 'imdb_id', right_on = 'imdb_id')
```

```python
merged_df
```

```python
# Another way of getting the above output
temp_df = imdb_data[(imdb_data['original_language'] == 'en') & (imdb_data['runtime'] >= 100)]
```

## Left Join to merge two df

```python
left_merged_df = pd.merge(english_movies,long_movies,how = 'left', left_on = 'imdb_id', right_on = 'imdb_id')
```

```python
left_merged_df.shape
```

## Right Join to merge two DF


```python
right_merged_df = pd.merge(english_movies,long_movies,how = 'right', left_on = 'imdb_id', right_on = 'imdb_id')
right_merged_df
```

```python
outer_join_df = pd.merge(english_movies,long_movies,how = 'outer',on='imdb_id')
outer_join_df
```
